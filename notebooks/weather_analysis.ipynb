```python
df['Date/Time'] = pd.to_datetime(df['Date/Time'], format='mixed', dayfirst=False)

df['Month'] = df['Date/Time'].dt.month
df['Hour'] = df['Date/Time'].dt.hour

print("Null values after fix:")
print(df[['Month', 'Hour']].isna().sum())
```

```python
df.drop(columns=['Date/Time'],inplace=True,axis=1)
```

```python
df.head()
```

```python
# Step 1: Split all columns that contain commas
for col in df.columns:
    if df[col].astype(str).str.contains(',').any():
        df[col] = df[col].astype(str).str.split(',')

# Step 2: Explode all list-like columns
for col in df.columns:
    if df[col].apply(lambda x: isinstance(x, list)).any():
        df = df.explode(col)

# Step 3: Clean up whitespace
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

print(df.reset_index(drop=True))
```

```python
df['Weather'].value_counts()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Define mapping logic
def simplify_weather(x):
    x = str(x).lower().strip()
    
    if 'clear' in x:  # Mainly Clear, Clear
        return 'Clear'
    elif 'cloud' in x or 'haze' in x:  # Cloudy, Mostly Cloudy, Haze
        return 'Cloudy'
    else:  # Everything else (Rain, Snow, Fog, Drizzle, etc.)
        return 'Bad Weather'

# Apply mapping
df['Weather Class'] = df['Weather'].apply(simplify_weather)

# Encode target labels
le = LabelEncoder()
df['Weather Class Encoded'] = le.fit_transform(df['Weather Class'])

# Show summary
print(df['Weather Class'].value_counts())
print(df.head())
```

```python
X = df[['Temp_C', 'Rel Hum_%', 'Wind Speed_km/h', 'Visibility_km', 'Press_kPa', 'Hour', 'Month']]
y = df['Weather Class']
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42)
X, y = sm.fit_resample(X, y)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from xgboost import XGBClassifier

model = XGBClassifier(
    n_estimators=500,
    learning_rate=0.05,
    max_depth=8,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)

model.fit(X_scaled, y)
```

```python
y_pred = model.predict(X_scaled)
```
